```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np

#fig = plt.figure(figsize=(8, 8))

'''
Germantown MWR Profile at 0200 UTC 30 June 2012
'''

tprof = 'mwr_germ_0630_0200_Tprof.txt'
rhprof = 'mwr_germ_0630_0200_RHprof.txt'
liqprof = 'mwr_germ_0630_Liqprof.txt'

T_press = np.loadtxt(tprof, usecols=range(1, 32), skiprows=1, unpack=True)
RH_press = np.loadtxt(rhprof, usecols=range(1, 32), skiprows=1, unpack=True)
LIQ_press = np.loadtxt(liqprof, usecols=range(1, 32), skiprows=1, unpack=True)
print("T_press = ", T_press)
print("T_press shape = ", T_press.shape)
print("RH_press = ", RH_press)
print("RH_press shape = ", RH_press.shape)
print("LIQ_press = ", LIQ_press)
print("LIQ_press shape = ", LIQ_press.shape)
#print("ILW = ", ILW_time)
#print("ILW shape = ", ILW_time.shape)
T = T_press[1:32,:]
T_K = T.T
RH = RH_press[1:32,:]
RH_pct = RH.T
LIQ = LIQ_press[1:32,:]
LIQ_g = LIQ.T
pressure_mb = T_press[0,:]
press = pressure_mb.tolist()
print("Pressure : ", press, len(press))
time = [120, 124, 128, 132, 136, 140, 144, 148, 152, 156, 160, 164, 168, 172, 176, 180, 184, 188, 192, 198, 202, 206, 210,
        214, 218, 222, 228, 232, 236, 240]
maxpress = max(press)
minpress = min(press)
maxtime = max(time)
mintime = min(time)
print("Latest time = ", maxtime)
timez = np.array([120, 124, 128, 132, 136, 140, 144, 148, 152, 156, 160, 164, 168, 172, 176, 180, 184, 188, 192, 198, 202, 206, 210,
        214, 218, 222, 228, 232, 236, 240])
print("Time: ", time, len(time))
print("Pressure in mb: ", pressure_mb, pressure_mb.shape)
print("Temperature in K: ", T_K, T_K.shape)
T_C = T_K - 273.15
timeutc, prlevs = np.meshgrid(timez, pressure_mb)
print("Time: ", timeutc, timeutc.shape)
print("Pr levels: ", prlevs, prlevs.shape)
TD_C = 243.04*((np.log(RH_pct/100))+((17.625*T_C)/(243.04+T_C)))/(17.625-(np.log(RH_pct/100))-((17.625*T_C)/(243.04+T_C)))
TD = TD_C + 273.15
WBT = T_C*(np.arctan(0.151977*((RH_pct+8.313659)**0.5))) + (np.arctan(T_C+RH_pct)) - (np.arctan(RH_pct-1.676331))+((0.00391838*(RH_pct**1.5))*(np.arctan(0.023101*RH_pct))) - 4.686035
Theta = T_K*((1000/prlevs)**0.286)
print("Theta in K: ", Theta, Theta.shape)
TH_min = np.amin(Theta)
TH_max = np.amax(Theta)
T_min = np.amin(T_K)
T_max = np.amax(T_K)
RH_min = np.amin(RH_pct)
RH_max = np.amax(RH_pct)
LIQ_min = np.amin(LIQ_g)
LIQ_max = np.amax(LIQ_g)
TD_C_min = np.amin(TD_C)
TD_C_max = np.amax(TD_C)
WBT_min = np.amin(WBT)
WBT_max = np.amax(WBT)
print("Theta min, Theta max = %f K %f K" % (TH_min, TH_max))
print("Temp min, Temp max = %f K %f K" % (T_min, T_max))
print("RH min, RH max = %f %f" % (RH_min, RH_max))
print("LIQ min, LIQ max = %f %f" % (LIQ_min, LIQ_max))
print("TD min, TD max = %f %f" % (TD_C_min, TD_C_max))
print("WBT min, WBT max = %f %f" % (WBT_min, WBT_max))

ThetaE = (273.15 + T_C)*((1000/prlevs)**0.286)+(3 * (RH_pct * (3.884266 * 10**
         ((7.5 * T_C)/(237.7 + T_C)))/100))
THe_sfc = ThetaE[1,:]
THe_top = ThetaE[20,:]
delta_zkm = 1.4
THe_min = np.amin(ThetaE)
THe_max = np.amax(ThetaE)
THe_mean = np.mean(ThetaE)
print("Thetae: ", ThetaE, ThetaE.shape)
print("Thetae min, Thetae max = %f K %f K" % (THe_min, THe_max))
THe_dev = ThetaE - THe_mean
THe_dev_min = np.amin(THe_dev)
THe_dev_max = np.amax(THe_dev)
print("Dev min, Dev max = %f K %f K" % (THe_dev_min, THe_dev_max))
levels_TH = np.arange(290,370,2)
levels_T = np.arange(230,310,2)
levels_RH = np.arange(35,100,2)
levels_LIQ = np.arange(0,10,0.1)
levels_TE = np.arange(315,410,2)
levels_Dev = np.arange(-25,70,0.5)
levels_WBT = np.arange(-40,35,0.5)
labels = ['0200', '', '', '', '0216', '', '', '', '0232', '', '', '', '0248', '', '', '0300', '', '', '0312',
          '', '', '', '0330', '', '', '', '0348', '', '', '0400']

gamma_ThetaE = (THe_sfc - THe_top)/(delta_zkm)
print("Gamma Theta-e = ", gamma_ThetaE)

fig = plt.figure(figsize=(12,12))
ax1 = fig.add_subplot(111)
im = ax1.imshow(T_K, extent =(mintime, maxtime, maxpress, minpress), aspect = 'auto', cmap = 'jet', interpolation = 'none', origin = 'lowest')
cb = plt.colorbar(im, label = 'K')
ax1.set_xlabel('Time (UTC)')
ax1.set_ylabel('Height (mb)')
plt.show()

fig = plt.figure(figsize=(12,12))
ax1 = fig.add_subplot(111)
im = ax1.imshow(RH_pct, extent =(mintime, maxtime, maxpress, minpress), aspect = 'auto', cmap = 'jet_r', interpolation = 'none', origin = 'lowest')
cb = plt.colorbar(im, label = '%')
ax1.set_xlabel('Time (UTC)')
ax1.set_ylabel('Height (mb)')
plt.show()

fig = plt.figure(figsize=(12,12))
ax1 = fig.add_subplot(111)
im = ax1.imshow(LIQ_g, extent =(mintime, maxtime, maxpress, minpress), aspect = 'auto', cmap = 'jet_r', interpolation = 'none', origin = 'lowest')
cb = plt.colorbar(im, label = '%')
ax1.set_xlabel('Time (UTC)')
ax1.set_ylabel('Height (mb)')
plt.show()

fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(111) # main axes
cf = ax.contourf(timeutc, prlevs, T_K, cmap=cm.jet, levels = levels_T)
plt.ylim(1000,270)
ax.set_xticks(time)
ax.set_xticklabels(labels)
plt.colorbar(cf)
plt.xlabel("Time (UTC)")
plt.ylabel("Height (mb)")
plt.title("Temperature (K) Germantown MD 30 June 2012")
plt.savefig("tempK_germ_0630.png",dpi=250,bbox_inches='tight')
plt.grid(True)
plt.show()

fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(111) # main axes
cf = ax.contourf(timeutc, prlevs, Theta, cmap=cm.jet, levels = levels_TH)
plt.ylim(1000,270)
ax.set_xticks(time)
ax.set_xticklabels(labels)
plt.colorbar(cf)
plt.xlabel("Time (UTC)")
plt.ylabel("Height (mb)")
plt.title("Theta (K) Germantown MD 30 June 2012")
plt.savefig("thta_germ_0630.png",dpi=250,bbox_inches='tight')
plt.grid(True)
plt.show()

fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(111) # main axes
cf = ax.contourf(timeutc, prlevs, ThetaE, cmap=cm.jet, levels = levels_TE)
plt.ylim(1000,270)
ax.set_xticks(time)
ax.set_xticklabels(labels)
plt.colorbar(cf)
plt.xlabel("Time (UTC)")
plt.ylabel("Height (mb)")
plt.title("Theta-E (K) Germantown MD 30 June 2012")
plt.savefig("thtae_germ_0630.png",dpi=250,bbox_inches='tight')
plt.grid(True)
plt.show()

fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(111) # main axes
cf = ax.contourf(timeutc, prlevs, THe_dev, cmap=cm.jet, levels = levels_Dev)
plt.ylim(1000,270)
ax.set_xticks(time)
ax.set_xticklabels(labels)
plt.colorbar(cf)
plt.xlabel("Time (UTC)")
plt.ylabel("Height (mb)")
plt.title("Theta-E Deviation (K) Germantown MD 30 June 2012")
plt.savefig("thtaedev_germ_0630.png",dpi=250,bbox_inches='tight')
plt.grid(True)
plt.show()

fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(111) # main axes
cf = ax.contourf(timeutc, prlevs, WBT, cmap=cm.jet, levels = levels_WBT)
plt.ylim(1000,270)
ax.set_xticks(time)
ax.set_xticklabels(labels)
plt.colorbar(cf)
plt.xlabel("Time (UTC)")
plt.ylabel("Height (mb)")
plt.title("Wet-bulb Temperature (C) Germantown MD 30 June 2012")
plt.savefig("wbt_germ_0630.png",dpi=250,bbox_inches='tight')
plt.grid(True)
plt.show()

fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(111) # main axes
cf = ax.contourf(timeutc, prlevs, LIQ_g, cmap=cm.jet, levels = levels_LIQ)
plt.ylim(1000,270)
ax.set_xticks(time)
ax.set_xticklabels(labels)
plt.colorbar(cf)
plt.xlabel("Time (UTC)")
plt.ylabel("Height (mb)")
plt.title("Liquid (g/m^3) Germantown MD 30 June 2012")
plt.savefig("liq_germ_0630.png",dpi=250,bbox_inches='tight')
plt.grid(True)
plt.show()

fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(111) # main axes
cf = ax1.contourf(timeutc, prlevs, LIQ_g, cmap=cm.jet, levels = levels_LIQ)
ax1.set_ylim(1000,270)
ax1.yaxis.tick_right()
ax1.yaxis.set_label_position("right")
ax1.set_xticks(time)
ax1.set_xticklabels(labels)
ax1.set_xlabel("Time (UTC)")
ax1.set_ylabel("Height (mb)")
ax1.set_title("Liquid (g/m^3) Germantown MD 30 June 2012")
ax2 = fig.add_subplot(111, sharex=ax1, frameon=False) # main axes
ax2.plot(minuteslwx,VIL,color='white',linewidth=5,label='DVIL LWX')
ax2.yaxis.tick_left()
ax2.yaxis.set_label_position("left")
ax2.set_ylabel("DVIL (kg/m^2)")
ax2.legend(loc='upper center')
# Adding the colorbar
cbaxes = fig.add_axes([0.99, 0.12, 0.03, 0.76]) 
cb = plt.colorbar(cf,cax = cbaxes)
plt.savefig("liq_dvil_germ_0630.png",dpi=250,bbox_inches='tight')
plt.show()

fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(111) # main axes
cf = ax1.contourf(timeutc, prlevs, LIQ_g, cmap=cm.jet, levels = levels_LIQ)
ax1.set_ylim(1000,270)
ax1.yaxis.tick_right()
ax1.yaxis.set_label_position("right")
ax1.set_xticks(time)
ax1.set_xticklabels(labels)
ax1.set_xlabel("Time (UTC)")
ax1.set_ylabel("Height (mb)")
ax1.set_title("Liquid (g/m^3) Germantown MD 30 June 2012")
ax2 = fig.add_subplot(111, sharex=ax1, frameon=False) # main axes
ax2.plot(minuteslwx,ILW,color='white',linewidth=5,label='MWR ILW')
ax2.yaxis.tick_left()
ax2.yaxis.set_label_position("left")
ax2.set_ylabel("ILW (mm)")
ax2.legend(loc='upper center')
# Adding the colorbar
cbaxes = fig.add_axes([0.99, 0.12, 0.03, 0.76]) 
cb = plt.colorbar(cf,cax = cbaxes)
plt.savefig("liq_ilw_germ_0630.png",dpi=250,bbox_inches='tight')
plt.show()

fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(111) # main axes
cf = ax1.contourf(timeutc, prlevs, Theta, cmap=cm.jet, levels = levels_TH)
ax1.set_ylim(1000,270)
ax1.yaxis.tick_right()
ax1.yaxis.set_label_position("right")
ax1.set_xticks(time)
ax1.set_xticklabels(labels)
ax1.set_xlabel("Time (UTC)")
ax1.set_ylabel("Height (mb)")
ax1.set_title("Theta (K) Germantown MD 30 June 2012")
ax2 = fig.add_subplot(111, sharex=ax1, frameon=False) # main axes
ax2.plot(minuteslwx,VIL,color='white',linewidth=5,label='DVIL LWX')
ax2.yaxis.tick_left()
ax2.yaxis.set_label_position("left")
ax2.set_ylabel("DVIL (kg/m^2)")
ax2.legend(loc='upper center')
# Adding the colorbar
cbaxes = fig.add_axes([0.99, 0.12, 0.03, 0.76]) 
cb = plt.colorbar(cf,cax = cbaxes)
plt.savefig("thta_dvil_germ_0630.png",dpi=250,bbox_inches='tight')
plt.show()

fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(111) # main axes
cf = ax1.contourf(timeutc, prlevs, ThetaE, cmap=cm.jet, levels = levels_TE)
ax1.set_ylim(1000,270)
ax1.yaxis.tick_right()
ax1.yaxis.set_label_position("right")
ax1.set_xticks(time)
ax1.set_xticklabels(labels)
ax1.set_xlabel("Time (UTC)")
ax1.set_ylabel("Height (mb)")
ax1.set_title("Theta-E (K) Germantown MD 30 June 2012")
ax2 = fig.add_subplot(111, sharex=ax1, frameon=False) # main axes
ax2.plot(minuteslwx,VIL,color='white',linewidth=5,label='DVIL LWX')
ax2.yaxis.tick_left()
ax2.yaxis.set_label_position("left")
ax2.set_ylabel("DVIL (kg/m^2)")
ax2.legend(loc='upper center')
# Adding the colorbar
cbaxes = fig.add_axes([0.99, 0.12, 0.03, 0.76]) 
cb = plt.colorbar(cf,cax = cbaxes)
plt.savefig("thtae_dvil_germ_0630.png",dpi=250,bbox_inches='tight')
plt.show()
```
